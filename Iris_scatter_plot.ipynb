```python
% matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('../iris/Iris.csv')
```

```python
 iris["Species"].value_counts()
```

```python
virginica = iris[iris.Species == "Iris-virginica"]
versicolor = iris[iris.Species == "Iris-versicolor"]
setosa = iris[iris.Species == "Iris-setosa"]
```

```python
versicolor.plot(x='SepalWidthCm',y='SepalLengthCm',color = 'blue',kind='scatter')
plt.title('Versicolor Iris sepal dimensions')
setosa.plot(x='SepalWidthCm',y='SepalLengthCm',color = 'green',kind='scatter')
plt.title('Setosa Iris sepal dimensions')
virginica.plot(x='SepalWidthCm',y='SepalLengthCm',color='red',kind='scatter')
plt.title('Virginica Iris sepal dimensions')
plt.show()
```
